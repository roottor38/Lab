### 전처리 과정

```python
import csv

with open("data/ratings.txt", encoding="utf-8-sig") as f:
    data = [r for r in csv.reader(f, delimiter='\t')][1:]
len(data)
```

```python
from konlpy.tag import Okt

okt = Okt()
```

```python
okt.pos(data[0][1])
```

```python
import json
import os

def tokenize(doc):
    # norm은 정규화, stem은 근어로 표시하기를 나타냄
    return ['/'.join(t) for t in okt.pos(doc, norm=True, stem=True)]

fname = 'processing.json'
if os.path.isfile(fname):
    with open(fname) as f:
        processing = json.load(f)
else:
    processing = [(tokenize(row[1]), row[2]) for row in data]
    # JSON 파일로 저장
    with open(fname, 'w', encoding="utf-8") as make_file:
        json.dump(processing, make_file, ensure_ascii=False, indent="\t")
```

```python
import nltk
from pprint import pprint

tokens = [t for d in processing for t in d[0]]
print(len(tokens))

text = nltk.Text(tokens, name='NMSC')

# 전체 토큰의 개수
print(len(text.tokens))

# 중복을 제외한 토큰의 개수
print(len(set(text.tokens)))            

# 출현 빈도가 높은 상위 토큰 10개
pprint(text.vocab().most_common(10))
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(processing)
df
```

```python
# 시간이 꽤 걸립니다! 시간을 절약하고 싶으면 most_common의 매개변수를 줄여보세요.
selected_words = [f[0] for f in text.vocab().most_common(10000)]

def term_frequency(doc):
    return [doc.count(word) for word in selected_words]
```

```python
t_freq = [term_frequency(d) for d in df[0]]
```

```python
bow_data = np.asarray(t_freq).astype('float32')
bow_label = df[1].astype('float32')
```

### 전처리 완료 => 딥러닝 진행

```python
from sklearn.model_selection import train_test_split
train_data, test_data, train_label, test_label = train_test_split(bow_data, bow_label)
```

```python
from keras.models import Sequential 
from keras.layers import Dense 
from keras import optimizers, losses, metrics
```

```python
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(train_data.shape[1],)))
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer=optimizers.RMSprop(lr=0.001),
             loss=losses.binary_crossentropy,
             metrics=[metrics.binary_accuracy])

model.fit(train_data, train_label, epochs=10, batch_size=512)
results = model.evaluate(test_data, test_label)
```

```python
results
```

```python
def predict_pos_neg(review):
    token = tokenize(review)
    tf = term_frequency(token)
    data = np.expand_dims(np.asarray(tf).astype('float32'), axis=0)
    score = float(model.predict(data))
    if(score > 0.5):
        print("[{}]는 {:.2f}% 확률로 긍정 리뷰이지 않을까 추측해봅니다.^^\n".format(review, score * 100))
    else:
        print("[{}]는 {:.2f}% 확률로 부정 리뷰이지 않을까 추측해봅니다.^^;\n".format(review, (1 - score) * 100))
```

```python
predict_pos_neg("이런 영화는 좋은 영화라고 하기 힘들지 않을까 싶어요")
```
